```python
import pandas
from matplotlib import pyplot
pandas.set_option("display.max_rows",2738913)
```

```python
path="F:\\Python\\project\\car.csv"
raw=pandas.read_csv(path,sep=',')
raw.columns=['buying_price','price_of_maintenance','nuber_of_door','person_capacity','size_of_luggage_boot','safety','car_acceptablility']
raw
```

```python
from sklearn import preprocessing
transform=preprocessing.LabelEncoder()
for a in raw.columns:
    raw[a]=transform.fit_transform(raw[a])
ans=raw['car_acceptablility']
raw=raw.drop('car_acceptablility',axis=1)
raw
```

```python
ans.value_counts(normalize=True)
```

```python
import numpy
from sklearn.neighbors import KNeighborsClassifier as KNN
data=numpy.array(raw)
quality=numpy.array(ans)
clf=KNN(n_neighbors=80)
clf.fit(data[100:],quality[100:])
clf.score(data[:100],quality[:100])
```
